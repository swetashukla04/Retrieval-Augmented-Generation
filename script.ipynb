```python
import pandas as pd
import requests
import pdfplumber
```

```python
pdf_path = "anual.pdf"
```

```python
data = []

with pdfplumber.open(pdf_path) as pdf:
    for i, page in enumerate(pdf.pages):
        text = page.extract_text()
        data.append({"page_number": i, "text": text})

data
```

```python
df = pd.DataFrame(data)
df.head()
```

# Vector Spaces
- Turn the docs into vectors
- Term-document matrix
 - rows: documents
 - columns: words/tokens

```python
from sklearn.feature_extraction.text import TfidfVectorizer
#TF-IDF stands for Term Frequency Inverse Document Frequency of records. It can be defined as the calculation of how relevant a word in a series or corpus is to a text
```

```python
cv = TfidfVectorizer(min_df=3, stop_words='english')
X = cv.fit_transform(df.text)
```

cv -> count vectorizer
X -> term-document matrix

```python
names = cv.get_feature_names_out() # Token stored
df_docs = pd.DataFrame(X.toarray(), columns=names) 
# df_docs.round(2)
df_docs.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
cv = TfidfVectorizer(stop_words='english', min_df=10)
X = cv.fit_transform(df.text)
```

```python
X
```

```python
query = "what is placement status of students"
q = cv.transform([query])
q.toarray()
```

```python
X.dot(q.T).todense() # this is the cosine similarity b/w query matrics and data matrics
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
score = cosine_similarity(X, q).flatten()
score
```

```python
import numpy as np

idx = np.argsort(score)[-5:]
```

```python
results = df.iloc[idx]
```

```python
results.to_dict(orient="records")
```

```python
from sklearn.decomposition import TruncatedSVD
# to compress the matric and reduce the dimentionality
```

```python
X
```

```python
svd = TruncatedSVD(n_components=16) # compressed into 16 components
X_emb = svd.fit_transform(X)
```

```python
X_emb[0]
```

```python
query = "what is placement status of students"
Q = cv.transform([query])
q_emb = svd.transform(Q)
q_emb[0]
```

```python
score = cosine_similarity(X_emb, q_emb).flatten()
idx = np.argsort(score)[-5:]
list(df.iloc[idx].text)
```

```python
from sklearn.decomposition import NMF

nmf = NMF(n_components=16)
W = nmf.fit_transform(X)
W[0]
```

```python
Q = cv.transform([query])
q_emb = nmf.transform(Q)
q_emb[0]
```

```python
score = cosine_similarity(W, q_emb).flatten()
idx = np.argsort(score)[-5:]
df.iloc[idx].text
```
